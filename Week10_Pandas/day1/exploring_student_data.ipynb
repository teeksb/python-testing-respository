```python
import pandas as pd
```

### Load the student_record_missing_values.csv file into a Pandas DataFrame

```python
students_df = pd.read_csv("../files/students_record_missing_values.csv")
```

### Explore the data structure (number of rows/columns, datat types)


```python
print("\n====================================== DATA EXPLORATION =================================================")
(rows, columns) = students_df.shape

print(f"\n{rows} rows, {columns} columns")
print("\n====================== DATA TYPES ========================================")
print(students_df.dtypes)


print("\n====================== MISSING VALUES ========================================")
print("\nMissing Values in each column")
print(students_df.isnull().sum())

print("\n====================== DATASET PREVIEW ========================================")
print(students_df.head())
```

```python
print(students_df.columns)
```

### Calculate basic statistics for numerical columns (mean, max, min, std dev)

```python

columns = students_df.columns
subject_cols = []

for col in columns:
    if "_score" in col:
        subject_cols.append(col)

stats_summary = students_df[subject_cols].describe()

print(students_df[subject_cols])
print("***********************************************")
print("***********************************************")
print("***********************************************")
print(stats_summary)
```

### Identify top 5 by average score across all subjects.

```python
students_df["average_score"] = students_df[subject_cols].mean(axis=1)  # axis -1 means across rows while axis-0 meand down columns(default)

top_students = students_df.sort_values("average_score", ascending=False)
print(top_students.head(5))
```

### Create a new column for "pass/fail" based on whether the average score is 75 

```python
def determine_pass_or_fail(average_score):
    return "Pass" if average_score >= 75 else "Fail"

    # You can also use the code below to return either "Pass" or "Fail" rather than booleans of True and False
    # return "Pass" if average_score >= 75 else "Fail"

students_df["pass_fail"] = students_df["average_score"].apply(determine_pass_or_fail)
print(students_df[["student_id", "average_score", "pass_fail"]])
```

### Find the pass percentage for entire school.

```python
pass_fail_counts = students_df["pass_fail"].value_counts()
print ("****")
print(pass_fail_counts)
print("")
print("********************************************")
number_of_passed_students = pass_fail_counts["Pass"]
print("Number of passed students:", number_of_passed_students)
pass_percentage = (number_of_passed_students/ len (students_df)) * 100
print(f"Pass percentage: {pass_percentage:.2f}%")

#The above can also be written as the below:
# print(f"\nPass Percentage: {pass_fail_counts["Pass"] / len(students_df) * 100}%")
```

```python
# def calc_percentage(series_groupby):
#     return (series_groupby == "Pass").mean() * 100
# # result = students_df.groupby("grade_level")["pass_fail"].apply(calc_pass_percentage)
# print(result)
#The above lines of code can simply be written as the below
result = students_df.groupby("grade_level")["pass_fail"].apply(lambda x: (x == "Pass").mean() * 100)

print(result)
```
